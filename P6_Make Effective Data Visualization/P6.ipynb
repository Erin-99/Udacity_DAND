```python
from datetime import * 

def convert_to_time(t):
    second=0
    if len(t)==4:
        hour=int(t[:2])
        minute=int(t[3:])
    elif len(t)>=3:
        hour=int(t[0])
        minute=int(t[1:])
    else:
        hour=0
        minute=int(t)
    new_time = time(hour,minute,second)
    return new_time
print convert_to_time('1800')
```
